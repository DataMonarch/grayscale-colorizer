```python
import imp
import os
import cv2
import pdb
import glob
import pandas as pd
import numpy as np
from tqdm import tqdm
from multiprocessing import Pool
import logging
import datetime
import sys
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("gray.jpeg")
plt.imshow(img)
```

```python
red = np.zeros((1, 1, 3), np.uint8)
red[:] = (0,0,255)

green = np.zeros((1, 1, 3), np.uint8)
green[:] = (0,255,0)

# create 1 pixel blue image
blue = np.zeros((1, 1, 3), np.uint8)
blue[:] = (255,0,0)

lut = np.concatenate((blue, red, green), axis=0)
print(lut)
# resize lut to 256 values
lut = cv2.resize(lut, (1,256), interpolation=cv2.INTER_CUBIC)

result = cv2.LUT(img, lut)
plt.imshow(result)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
